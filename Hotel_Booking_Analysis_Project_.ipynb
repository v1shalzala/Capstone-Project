#### Missing Values/Null Values

```python
# Missing Values/Null Values Count
hotel_dataset.isnull().sum()
```

```python
# Visualizing the missing values
s = hotel_dataset.isnull().sum()
df = pd.DataFrame({'column_name':s.index, 'null_count':s.values})
plt.figure(figsize=(10,7))
ax = sns.barplot(x="column_name",y='null_count',data=df)

ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
plt.tight_layout()
plt.show()
```

As of the above results we get to see that the columns such as agent , company and country have high count of null values which will impact on our future analyis.

**Solution** : There are only two solutions to above problem one is to drop the highest null value count columns and second way is that we can fill those null values with some value which is an appropirate way to manipulate the data where there is no need to drop the column.

#2. **DATA CLEANING** / **FILLING MISSING VALUES**



Now we will fill the null values by adding some values so that they do not cause any further problem in the analysis.
###Columns that should be gone in data cleaning are :

1.   agent (most priority)
2.   company (most priority)
3.   country (most priority)
4. children



```python
# Replacing NaN values to 0.0 in agent and company columns values
hotel_dataset[['agent','company']] = hotel_dataset[['agent','company']].fillna(0.0)
```

```python
#after filling null values with 0.0 count will become zero
hotel_dataset.isnull().sum()
```

Now we only have two columns (children and country) that are to be either droped or fillied with some values

```python
hotel_dataset['country'].fillna(hotel_dataset['country'].mode().to_string(), inplace=True)
hotel_dataset['children'].fillna(round(hotel_dataset['children'].mean()), inplace=True)
```



*   We successfully cleaned the dataset and before proceeding further we check the count of the null values



```python
hotel_dataset.isnull().sum()
```

```python
j = hotel_dataset.isnull().sum()
df3 = pd.DataFrame({'column_name':j.index, 'null_count':j.values})
plt.figure(figsize=(10,7))
ax = sns.barplot(x="column_name",y='null_count',data=df3)

ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
plt.tight_layout()
plt.show()
```

Above visualization is the result of the entire data cleaning process where no NaN values are present in the dataset

##### 1. Why did you pick the specific chart?

Answer - Pie charts provide a clear and immediate comparison between canceled and not cnceled bookings.

Pie charts inherently display data as percentages of the whole, making it easy to see what percentage of bookings fall into each category. This can be valuable for understanding the overall picture.

##### 2. What is/are the insight(s) found from the chart?
